```python
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup

print('Libraries imported')
```

1- Get the desired web page from URL

2- Access the content (HTML) of this web page via BeautifulSoup

3- Find the desired tag in the html content (in this case the table "wikitable")

4- Transform the Soup object into a pandas data frame

```python
url = 'http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# Get the desired web page
data = requests.get(url)

# Access the content (HTML) of this web page via BeautifulSoup
soup = BeautifulSoup(data.content,'lxml')
#print(soup.prettify())

# Find the desired tag in the html content
match = soup.find('table', class_='wikitable')
#print(match)

# Transform the Soup object into a pandas data frame
df = pd.read_html(str(match))[0]
df.shape
```

```python
df.head(15)
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
condition = df[df['Borough']=='Not assigned'].index
df.drop(condition, inplace = True)
df.reset_index(drop = True, inplace = True)
df.head(10)
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df['Neighborhood'].replace(np.nan, df['Borough'], inplace = True)
df.head(10)
```

```python
df.shape
```

Import the csv file and adjust the columns name

```python
csv = 'http://cocl.us/Geospatial_data'
gl = pd.read_csv(csv)
gl.columns=["Postal code", "Latitude", "Longitude"]
gl.head()
```

Merge the two data frames based on the column Postal code

```python
result = pd.merge(df,gl, on='Postal code')
result.head(50)
```

```python
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium
```

Extract the boroughs that contain the word Toronto

```python
Toronto = result[result['Borough'].str.contains('Toronto')]
Toronto.head()
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(Toronto['Latitude'], Toronto['Longitude'], Toronto['Borough'], Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
